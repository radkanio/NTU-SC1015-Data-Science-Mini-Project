## Combine the CSV Files

The dataset only contained monthly data per year. We decided to combine each month's CSV file so that it gives us a new CSV file which contains the whole data for 2021. (Note: We originally used this code in a different notebook so we did not technically run it here, but the code still works if you downloaded the right CSV files)

```python
import pandas as pd
import glob

# Path to the directory containing the CSV files
directory = '/path/to/csv/files/'

# List all CSV files in the directory
files = glob.glob(directory + 'Flights_2021_*.csv')

# Initialize an empty list to store DataFrame objects
dfs = []

# Read each CSV file and append its DataFrame to the list
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Write the combined DataFrame to a new CSV file
combined_df.to_csv('/path/to/output/combined_flights_2021.csv', index=False)
```

## Basic Cleaning

We observe a column called 'Duplicate'. We drop such duplicates. 
We also drop the column. 

```python
flight_data = flight_data[flight_data['Duplicate'] != 'Y'].drop_duplicates()

flight_data = flight_data.drop('Duplicate', axis=1)

flight_data.info(verbose=True)
```

We drop the 'Div' columns (from column 74 and onwards) because we are not in interest of diverted flights, but more on delayed ones.

```python
flight_data = flight_data.iloc[:, :73]
flight_data.info(verbose=True)
```

We now tackle NA entries. We compute the percentage of data entries for each column that are not 100%.

```python
# Count non-null entries in each column until column 72
entry_counts = flight_data.iloc[:, :73].count()

# Total number of rows in the DataFrame
total_rows = len(flight_data)

# Compute the percentage of data entries for each column
entry_percentages = (entry_counts / total_rows) * 100

# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Filter the percentages where the entry percentage is not 100
filtered_percentages = entry_percentages[entry_percentages < 100]

# Display the result
print("Percentage of data entries for each column that are not 100%:")
print(filtered_percentages)
```

Let us focus more on the columns with significantly less entries, say 2%.
We shall drop such ones.

```python
# Filter columns with less than 2% data entries
columns_to_drop = entry_percentages[entry_percentages < 2].index

# Drop columns with less than 2% data entries
flight_data = flight_data.drop(columns_to_drop, axis=1)

# Print information about the DataFrame
flight_data.info(verbose=True)
```

Lets proceed to understanding the data columns. See how different 'DepDelayMinutes' and 'DepDelay' are different.

We deduce that DepDelayMinutes is 0 when the flights did not delay (i.e. DepDelay is negative). Continue exploring.

We deduce that DepDel15 indicates delays. If the flights delay, it will show 1 (see row 36), and 0 if not.  

Lets understand what 'DepartureDelayGroups' is next. 

```python
unique_groups = flight_data['DepartureDelayGroups'].unique()

# Print unique groups
print("Unique groups of DepartureDelayGroups:")
for group in sorted(unique_groups):
    print(group)
```

```python
# Group by 'DepartureDelayGroups' and get the first occurrence of 'DepDelay' within each group
first_occurrences = flight_data.groupby('DepartureDelayGroups')['DepDelay'].first()

# Print the first occurrences
print("First occurrence of DepDelay corresponding to each DepartureDelayGroup:")
for group, dep_delay in first_occurrences.items():
    print(f"Group {group}: {dep_delay}")
```

We find that DepartureDelayGroup shows the intervals of delays. Each increment of 1 for DepartureDelayGroup

Lets now see how diverted flights are treated. 

```python
# Count number of diverted flights
diverted_count = flight_data[flight_data['Diverted'] == True]['Diverted'].count()
print("Number of diverted flights:", diverted_count)

# Count number of cancelled flights
cancelled_count = flight_data[flight_data['Cancelled'] == True]['Cancelled'].count()
print("Number of cancelled flights:", cancelled_count)

# Count number of flights that are both diverted and cancelled
both_diverted_cancelled_count = flight_data[(flight_data['Diverted'] == True) & (flight_data['Cancelled'] == True)].shape[0]
print("Number of flights that are both diverted and cancelled:", both_diverted_cancelled_count)
```

We read that diverted flights are considered legally cancelled (source: https://www.iuno.law/news/consequences-of-a-diverted-flight/#:~:text=Thus%2C%20if%20a%20flight%20has,the%20flight%20had%20been%20delayed), so we had to see if one was the subset of the other.

We thus see that diverted flights are counted differently than cancelled flights for this dataset.

Lets see more on the airports. We will install and use a library in the name of 'pyairports'.

```python
from pyairports.airports import Airports, AirportNotFoundException

# Retrieve unique origins
origin_unique = flight_data['Origin'].unique()

# Create an empty list to store airport information
airport_info_list = []

# Initialize Airports object
airports = Airports()

# Retrieve airport information for each origin
for origin in origin_unique:
    try:
        airport_info = airports.lookup(origin)
        airport_info_list.append(airport_info)
    except AirportNotFoundException as e:
        print(f"Airport information not found for {origin}: {e}")

# Create a DataFrame with the airport information
data = {
    'iata': [airport_info.iata for airport_info in airport_info_list],
    'latitude': [float(airport_info.lat) for airport_info in airport_info_list],
    'longitude': [float(airport_info.lon) for airport_info in airport_info_list],
}

df = pd.DataFrame(data)
print(df)
```

```python
# Retrieve unique destinations
dest_unique = flight_data['Dest'].unique()

# Create an empty list to store airport information
dest_airport_info_list = []

# Initialize Airports object
airports = Airports()

# Retrieve airport information for each destination
for dest in dest_unique:
    try:
        airport_info = airports.lookup(dest)
        dest_airport_info_list.append(airport_info)
    except AirportNotFoundException as e:
        print(f"Airport information not found for {dest}: {e}")

# Create a DataFrame with the airport information for destinations
dest_data = {
    'iata_dest': [airport_info.iata for airport_info in dest_airport_info_list],
    'latitude_dest': [float(airport_info.lat) for airport_info in dest_airport_info_list],
    'longitude_dest': [float(airport_info.lon) for airport_info in dest_airport_info_list],
}

dest_df = pd.DataFrame(dest_data)
print(dest_df)
```

We observe that for some airports, the information was not found. We see if we can drop rows which contain these.

```python
# Define the list of values to check for in the 'Origin' and 'Dest' columns
values_to_check = ['USA', 'OGD', 'SWO', 'XWA', 'BIH']

# Count the number of rows where 'Origin' or 'Dest' contains any of the specified values
num_rows = ((flight_data['Origin'].isin(values_to_check)) | (flight_data['Dest'].isin(values_to_check))).sum()

print("Number of rows with Origin or Dest containing the specified values:", num_rows)

# Print the total number of rows in the flight_data DataFrame
print("Total number of rows in flight_data:", len(flight_data))

print("The percetage of such rows out of all the data is:", 5914/6185870)
```

We proceed to drop it.

```python
# Create boolean masks for rows to keep (False for rows with values to drop)
mask_origin = flight_data['Origin'].isin(values_to_check)
mask_dest = flight_data['Dest'].isin(values_to_check)
mask_keep = ~(mask_origin | mask_dest)

# Create a new DataFrame without rows containing specified values in 'Origin' or 'Dest' columns
cleaned_flight_data = flight_data[mask_keep]

# Print the total number of rows in the cleaned_flight_data DataFrame
print("Total number of rows in cleaned_flight_data:", len(cleaned_flight_data))
```
